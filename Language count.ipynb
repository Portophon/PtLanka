```python
import pandas as pd
import numpy as np
```

## Load data set

```python
############# mudar dados de entrada !!!!!!!!!!!!!!!!!!
languages_df = pd.read_csv ('languages.csv')
languages_df
```

```python
# Stats
```

```python
languages_df = languages_df.assign(Related_languages=np.core.defchararray.split(languages_df.Related_languages.values.astype(str), ','))
languages_df
```

```python
languages_list = list(languages_df["Related_languages"])
languages_list
```

```python
languages_list = [item for sublist in languages_list for item in sublist]
languages_list = [i.strip() for i in languages_list]
#languages_list = list(map(lambda item: item.replace("nan","Portuguese"), languages_list))
languages_list
```

```python
df = pd.DataFrame(languages_list, columns=["languages"]) 
df
```

```python
df_count = df['languages'].value_counts().rename_axis('languages').reset_index(name='counts')
df_count
```

```python
df_count.to_csv("language_count.csv", index = False, header = True)
```
